```python
# 2017

driver = webdriver.Chrome()
# all_links = []
links = []

ref = 'https://www.imdb.com/search/keyword/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=bdc91cb7-0144-4906-b072-b45760c8aa67&pf_rd_r=PXCPH97MT3PHQD6DF4AY&pf_rd_s=right-1&pf_rd_t=15051&pf_rd_i=genre&ref_=kw_ref_yr&mode=detail&'
prod = '&title_type=movie&release_date=1989%2C1989&sort=moviemeter,asc'
for i in range(21,41,1):
    time.sleep(random.uniform(2,4))
    url = ref + 'page=' + str(i) + prod
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    divs = soup.find('div', {'class':'lister-list'}).find_all('div', {'class':'lister-item mode-detail'})
    for div in divs:
        a = div.find('a').get('href')
        links.append(a)
#     all_links.append(links)
```

```python
data_for_all_links = pd.DataFrame([])
# all_links_ser = pd.Series(a)
# len_list = range(0,len(all_links_ser))
link = 'https://www.imdb.com'

for ref in links:
    
#     data_for_some_links = pd.DataFrame([])
#     for j in i:
#         print(j)
    url = link + ref
    print(url)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    try:
        data_for_dir = soup.find('div', {'class':'credit_summary_item'})
        a = data_for_dir.find('a')
        director = list(a)[0]
    except AttributeError:
        director = nan
    except IndexError:
        director = nan

    print('director', director)
    
    # 
    
    try:
        div = soup.find('div', {'id':'titleStoryLine'})
        divs = div.find_all('div',{'class':'see-more inline canwrap'})
        a = len(list(divs))
        div = list(divs)[a-1]
        aaa = div.find_all('a')
        genres = []
        for i in aaa:
#             print(i)
            k = list(i)[0]
            genres.append(k)
        len_genres = len(genres)
    except AttributeError:
        genres = nan
        len_genres = nan
    except IndexError:
        genres = nan
        len_genres = nan

    print('genres', genres)
    
    #  
    
    try:
        div = soup.find('div', {'id':'titleStoryLine'})
        divs = div.find_all('div',{'class':'see-more inline canwrap'})
        key_word = divs[0].text.strip().split('\n')[1]
    
    except AttributeError:
        key_word = nan
    except IndexError:
        key_word = nan
    
    # 
    
    try:
        div = soup.find('div', {'class':'plot_summary'}).contents[5]
        gen = div.find_all('a')
        writer = list(gen[0])[0]
    #     writers = []
    #     for i in gen:
    #         writers.append(i.text)
    except AttributeError:
        writer = nan
    except IndexError:
        writer = nan

    print('writer', writer)

    # 
    
    try:
        name = soup.find('div', {'class':'originalTitle'}).text
    except AttributeError:
        try:
            divs = soup.find('div', {'class':'title_wrapper'})
            div = soup.find('h1', {'class':''})
            name = div.contents[0].string
        except IndexError:
            name = nan
        except AttributeError:
            name = nan      
    except IndexError:
        name = nan
        
    print('name', name)
    
    #  
    
    try:
        divs = soup.find('div', {'class':'title_wrapper'})
        div = divs.find_all('a')
        a = len(div) - 1
        b = list(div[a])
        date_release = b[0]
#         print(b)
#         print(date_release)
    except IndexError:
        date_release = nan
    except AttributeError:
        date_release = nan

    print('date_release', date_release)
    
    # 
    
    try:
        budget = nan
        div = soup.find('div', {'id':'titleDetails'})
        aa = div.find_all('div',{'class':"txt-block"})
        for j in range(len(aa)):
            if aa[j].text.strip().split('\n')[0][:7] == 'Budget:':
                num = j
                print(num)
                budget = list(aa[j])[2]     
            else:
                continue 
    except AttributeError:
        budget = nan
    except IndexError:
        budget = nan    

    print('budget', budget)
    
    #      ( )    
    
    try:
        opening_weekend_usa = nan
        div = soup.find('div', {'id':'titleDetails'})
        aa = div.find_all('div',{'class':"txt-block"})
        for j in range(len(aa)):
            if aa[j].text.strip().split('\n')[0][:20] == 'Opening Weekend USA:':
                num = j
                print(num)
                opening_weekend_usa = list(aa[j])[2]     
            else:
                continue
    except AttributeError:
        opening_weekend_usa = nan
    except IndexError:
        opening_weekend_usa = nan
    
    print('opening_weekend_usa', opening_weekend_usa)
    
    #    
    
    try:
        gross_usa = nan
        div = soup.find('div', {'id':'titleDetails'})
        aa = div.find_all('div',{'class':"txt-block"})
        for j in range(len(aa)):
            if aa[j].text.strip().split('\n')[0][:10] == 'Gross USA:':
                num = j
                print(num)
                gross_usa = list(aa[j])[2]     
            else:
                continue
    except AttributeError:
        gross_usa = nan
    except IndexError:
        gross_usa = nan    

    print('gross_usa', gross_usa)
    
    #    
    
    try:
        cumulative_worldwide_gross = nan
        div = soup.find('div', {'id':'titleDetails'})
        aa = div.find_all('div',{'class':"txt-block"})
        for j in range(len(aa)):
            if aa[j].text.strip().split('\n')[0][:27] == 'Cumulative Worldwide Gross:':
                num = j
                print(num)
                cumulative_worldwide_gross = list(aa[j])[2]     
            else:
                continue
                cumulative_worldwide_gross = nan
    except AttributeError:
        cumulative_worldwide_gross = nan
    except IndexError:
        cumulative_worldwide_gross = nan    
    
    print('cumulative_worldwide_gross', cumulative_worldwide_gross)
    
    #  
    
    try:
        div = soup.find('div', {'id':'titleDetails'})
        aa = div.find_all('div',{'class':"txt-block"})
        for j in range(len(aa)):
            if aa[j].text.strip().split('\n')[0] == 'Country:':
                num = j
                aaa = aa[num].find_all('a')
                l = len(aaa)
                country = []
                for i in aaa[:l]:
                    p = list(i)[0]
                    print(p)
                    country.append(p)
                len_country = len(country)
            else:
                continue
                country = nan
                len_country = 0
            
    except AttributeError:
        country = nan
        len_country = 0
        
    except IndexError:
        country = nan
        len_country = 0
        
    print('country', country)
    
    #  
    
    try:
        production_co = nan
        len_production_co = 0
        div = soup.find('div', {'id':'titleDetails'})
        aa = div.find_all('div',{'class':"txt-block"})
        for j in range(len(aa)):
            if aa[j].text.strip().split('\n')[0] == 'Production Co:':
                num = j
                aaa = aa[num].find_all('a')
                l = len(aaa)-1
                production_co = []
                for i in aaa[:l]:
                    p = list(i)[0]
                    print(p)
                    production_co.append(p)
                len_production_co = len(production_co)
            else:
                continue
        
    except AttributeError:
        production_co = nan
        len_production_co = 0
        
    except IndexError:
        production_co = nan
        len_production_co = 0
    
    print('production_co', production_co)

    #  ()
    try:
        div = soup.find('div', {'id':'titleStoryLine'})
        gen = div.find_all('div', {'class':'inline canwrap'})
        data = []
        for i in gen:
            a = i.find_all('span')
            data.append(a)
        description = list(data[0][0])[0]
    except IndexError:
        description = nan
    except AttributeError:
        description = nan

    # 
    
    try:
        divs = soup.find('div', {'class':'title_wrapper'})
        div = divs.find('time', {})
        run_time = div.contents[0].string
    except IndexError:
        run_time = nan
    except AttributeError:
        run_time = nan  


    #  
    
    try:
        divs = soup.find('div', {'class':'title_wrapper'}) #   
        div = divs.find('div', {'class':'subtext'})
        age_rating = div.contents[0]
    except IndexError:
        age_rating = nan
    except AttributeError:
        age_rating = nan

    # 
    
    try:
        div = soup.find('div', {'class':'titleReviewBarItem'}) #   (,   )
        meta = div.find_all('span')
        list_for_meta = []
        for i in meta:
            list_for_meta.append(i.string)
        metascore = list_for_meta[0]
    except IndexError:
        metascore = nan
    except AttributeError:
        metascore = nan
        
    print('metascore', metascore)
    
    #  IMdb
    try:
        div = soup.find('div', {'class':'imdbRating'}) #  IMdb, -   . 
        imdb_rat = div.find_all('span')
        list_for_imdb_rat = []
        for i in imdb_rat:
            list_for_imdb_rat.append(i.string)
        imdb_rating = list_for_imdb_rat[0]
    except IndexError:
        imdb_rating = nan
    except AttributeError:
        imdb_rating = nan


    #  
    try:
        div = soup.find('div', {'class':'imdbRating'}) #  IMdb, -   . 
        imdb_rat = div.find_all('span')
        list_for_imdb_rat = []
        for i in imdb_rat:
            list_for_imdb_rat.append(i.string)
        all_who_put_rating = list_for_imdb_rat[3]
    except IndexError:
        all_who_put_rating = nan
    except AttributeError:
        all_who_put_rating = nan    

    #  ,   
    try:
        div = soup.find('div', {'class':'imdbRating'}) #  IMdb, -   . 
        imdb_rat = div.find_all('span')
        list_for_imdb_rat = []
        for i in imdb_rat:
            list_for_imdb_rat.append(i.string)
        all_users_who_put_rewiew = list_for_imdb_rat[4]
    except IndexError:
        all_users_who_put_rewiew = nan
    except AttributeError:
        all_users_who_put_rewiew = nan
    
    print('all_users_who_put_rewiew', all_users_who_put_rewiew)


    #  ,  
    
    try:
        div = soup.find('div', {'class':'imdbRating'}) #  IMdb, -   . 
        imdb_rat = div.find_all('span')
        list_for_imdb_rat = []
        for i in imdb_rat:
            list_for_imdb_rat.append(i.string)
        all_critics_who_put_rewiew = list_for_imdb_rat[5]
    except IndexError:
        all_critics_who_put_rewiew = nan
    except AttributeError:
        all_critics_who_put_rewiew = nan

    print('all_critics_who_put_rewiew', all_critics_who_put_rewiew)


    data = (
        [ref, name, genres, len_genres, key_word, director, writer,
         date_release, country, len_country, opening_weekend_usa, budget, gross_usa, 
         cumulative_worldwide_gross, production_co, len_production_co, 
         run_time, description, age_rating,
         metascore, all_users_who_put_rewiew,
         all_critics_who_put_rewiew, imdb_rating, all_who_put_rating
        ]
    )
    
    print(data)

    col = (
        ['Unique_link','Name','Genres', 'Len_genres', 'Key_word', 'Director','Writer','Date_release', 
         'Country', 'Len_country',
         'Opening_week_usa',
         'Budget', 'Gross',
         'Cum_world_gross','Production_co', 'Len_production_co',
         'Run_time','Description',
         'Age_rating',
         'Metascore','Number_of_rewiews',
         'Number_of_critics','Imdb_rating','Number_of_who_rated']
    )
    
    total_data = (
    (pd.DataFrame(
    data,col
    )).transpose()
    )

    print(total_data)

    data_for_all_links = data_for_all_links.append(total_data)
    
#     data_for_some_links = (data_for_all_links).drop_duplicates()
    data_for_all_links.to_excel('C:/Users/1/Documents/Project/New_data/1989_1000_2000_est.xlsx', index = False)

#         total_data = total_data.append(total_data)

#         data_for_some_links = (data_for_some_links.append(total_data)).drop_duplicates()

#     data_for_all_links = data_for_all_links.append(data_for_some_links)
```
